```python
def extract(filename):
    text_file = open(filename, "r")
    data = text_file.read()
    text_file.close()
    # print(data)
    ds = data.splitlines()
    for i in range(len(ds)):
        ds[i] = ds[i].split()
    # print(ds)
    return ds
```

```python
def multiple_fullstops(data_split):
    l2 = []
    for i in data_split:
        for j in i:
            if '...' in j:
                x = list(j)
                tmp = ''
                l1 = []
                for k in x:
                    if k != '.':
                        tmp += k
                    else:
                        break
                l1.append(tmp)
                tmp = ''
                for k in x:
                    if k == '.':
                        tmp += k
                l1.append(tmp)
                tmp2 = ''.join(l1)
                tmp = j.split(tmp2, 1)[1]
                if tmp != '':
                    l1.append(tmp)
                l2.append([j, l1])

    for i in range(len(data_split)):
        for j in range(len(data_split[i])):
            for k in l2:
                if data_split[i][j] == k[0]:
                    data_split[i][j] = k[1]
    ds2 = []
    for i in data_split:
        tmp = []
        for j in i:
            if isinstance(j, list):
                for k in j:
                    tmp.append(k)
            else:
                tmp.append(j)
        ds2.append(tmp)
    ds3 = [x for x in ds2 if x != []]
    return ds3
```

```python
def combine_spacedWords(ds):
    l = []
    l3 = []
    for i in range(len(ds)):
        tmp = []
        for j in range(len(ds[i])):
            if len(ds[i][j]) == 1 and ds[i][j].isalpha() and (ds[i][j].lower() != 'a') and (ds[i][j].lower() != 'i'):
                tmp.append(ds[i][j])
        l.append(tmp)
    l2 = [x for x in l if x != []]
    for i in l2:
        if len(i) > 1:
            l3.append(i)
    l2 = []
    for i in l3:
        x = ''.join(i)
        l2.append([i, x])

    for i in range(len(ds)):
        for j in range(len(ds[i])):
            for k in l2:
                if all(item in ds[i] for item in k[0]):
                    ds[i][j] = k[1]
                    ds[i] = [i for i in ds[i] if i not in k[0]]
    return ds
```

```python
def split_multiword(ds):
    l = []
    apple = ['ios', 'iphone', 'ipad', 'iphones', 'ipads']
    for i in range(len(ds)):
        for j in range(len(ds[i])):
            if any(l.isupper() for l in ds[i][j]) and ds[i][j].isalpha() and ds[i][j][0].islower() and ds[i][j].lower() not in apple:
                l.append(ds[i][j])
    l = list(set(l))
    l2 = []
    for i in l:
        x = list(i)
        tmp = ''
        l3 = []
        for j in x:
            if j.islower():
                tmp += j
            else:
                break
        l3.append(tmp)
        tmp = ''
        for j in range(len(x)):
            if x[j].isupper():
                tmp = x[j:]
        t = ''.join(tmp)
        l3.append(t)
        l2.append([i, l3])

    for i in range(len(ds)):
        for j in range(len(ds[i])):
            for k in l2:
                if ds[i][j] == k[0]:
                    ds[i][j] = k[1]
    ds3 = []
    for i in ds:
        tmp = []
        for j in i:
            if isinstance(j, list):
                for k in j:
                    tmp.append(k)
            else:
                tmp.append(j)
        ds3.append(tmp)
    return ds3
```

```python
def handle_fullstops(ds):
    l = []
    l2 = []
    l3 = []
    for i in ds:
        for j in i:
            if '.' in j and 'http' not in j and '.com' not in j and 'twitter' not in j and j != 'a.m.' and j != 'p.m.' and '..' not in j and j != '.':
                l.append(j)
    for i in l:
        x = i.split('.')
        if not ''.join(x).isdigit():
            y = list(i)
            nd = 0
            na = 0
            for j in range(len(y)):
                if y[j] == '.':
                    nd+=1
                else:
                    na+=1
            if nd == na or nd == na - 1:
                continue
            else:
                l2.append(i)
    for i in l2:
        x = i.split('.')
        tmp = []
        for k in x:
            if k != '':
                tmp.append(k)
                tmp.append('.')
        l3.append([i, tmp])

    for i in range(len(ds)):
        for j in range(len(ds[i])):
            for k in l3:
                if ds[i][j] == k[0]:
                    ds[i][j] = k[1]

    ds3 = []
    for i in ds:
        tmp = []
        for j in i:
            if isinstance(j, list):
                for k in j:
                    tmp.append(k)
            else:
                tmp.append(j)
        ds3.append(tmp)
    return ds3
```

```python
def handle_hyphens(ds):
    l = []
    for i in ds:
        for j in i:
            if len(j) != 1 and '-' in j and 'CF:D:' not in j and 'CF:T:' not in j and j[0] != '-' and j[0] != '@' and j[0] != '#' and 'http' not in j and '.com' not in j and 'twitter' not in j:
                l.append(j)
    l = list(set(l))
    l2 = []
    for i in l:
        x = i.split('-')
        tmp = []
        tmp.append(x[0])
        for k in range(1, len(x)):
            if not x[k][:1].isupper():
                tmp.append('-' + x[k])
            else:
                tmp.append('-')
                tmp.append(x[k])
        l2. append([i, tmp])

    for i in range(len(ds)):
        for j in range(len(ds[i])):
            for k in l2:
                if ds[i][j] == k[0]:
                    ds[i][j] = k[1]
    ds3 = []
    for i in ds:
        tmp = []
        for j in i:
            if isinstance(j, list):
                for k in j:
                    tmp.append(k)
            else:
                tmp.append(j)
        ds3.append(tmp)
    return ds3
```
